## Question:2

###  How to stack two arrays vertically?

#### Desired Output::

```python
import numpy as np
a1=np.arange(10).reshape(2,5)
a2=np.ones(10, dtype=int).reshape(2,5)
q2=np.vstack(a1,a2)
```
